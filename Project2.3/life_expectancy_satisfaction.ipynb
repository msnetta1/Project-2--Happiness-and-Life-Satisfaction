```python
import pandas as pd
from sqlalchemy import create_engine



csv_file = "./db/data.csv"
life_satifaction = pd.read_csv(csv_file)
life_satifaction.head()



database_path = "life_expectancy_satisfaction"
engine = create_engine(f"sqlite:///{database_path}")



engine.table_names()


life_satifaction.to_sql(name='life_expectancy_satisfaction', con=engine, if_exists='append', index=False)




pd.read_sql_query('select * from life_expectancy_satisfaction', con=engine)
```
